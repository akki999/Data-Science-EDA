####  1.7 importing import skew, norm

```python
from scipy.stats import norm,skew
```

```python
sns.kdeplot(wine['quality'])
plt.show()
```

#### 1.8 Create the numeric_features to remove skewness

```python
wine['sqrt'] = np.sqrt(wine['quality'])
wine['log'] = np.log(wine['quality'])
mu = wine['quality'].mean()
variance = wine['quality'].var()
sigma = np.sqrt(variance)
x = np.linspace(mu-3*variance,mu+3*variance, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.show()
```

#### 1.9 Create the skewed_features = wines[numeric_features]

```python
wine.head()
```

#### 1.10 Create DataFrame for skewness?

```python
skewness = pd.DataFrame(data=wine.skew()).rename(columns={0:'Value'}).reset_index().rename(columns={'index':'parameter'})
skewness
```

#### 1.11 Create to Get only higest skewed = [abs(skewness) > 0.7] features?

```python
skewness[(abs(skewness['Value'])>0.7)]
```

```python
# Get only higest skewed features
skewness[(abs(skewness['Value'])>0.7)]['parameter']
```

#### 1.12 Create skewness.index and new skewed_features2

```python
#Create Empty list 
skewed_features2 = []
#create for loop for skewness.index
for i in skewness.index:
    skewed_features2.append(skewness.loc[i])
#create skewed_features2 for skewness.index
skewed_features2
```

#### 1.13 Create new DataFrame for skewed_features2?

```python
#compute skewness
skewness2 = pd.DataFrame(data=skewed_features2)
# concat both skewness, skewness2
skew = skewness+skewness2
skew
```
